# Replace some substrings

```python

new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^УЛ. ', 'УЛ.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'Добрич - град', 'Добрич',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^УЛИЦ', 'УЛ.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^ЖК.', 'Ж.К.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^Ж.К. ', 'Ж.К.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^КВ. ', 'КВ.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^ВЗ', 'В.З.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^БЛ. ', 'БЛ.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^С. ', 'С.',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'^NO', ' NO',regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r' П.\d+','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r' АП.*','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r' ЕТ.*','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r' ВХ.*','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r' БЛ.*','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'NO.*','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'N.*','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'Ж.К ИЗТОК','Ж.К.ИЗТОК', regex=True)

new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'\s[0-9]+,$','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'\s-[0-9]+,$','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'\s[0-9]+$','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'\s[А-Яа-яЁё]$','', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r'БУЛ .*','БУЛ.*', regex=True)
new_nsi_df['PA_test'] = new_nsi_df['PA_test'].str.replace(r',.*','', regex=True)
```
