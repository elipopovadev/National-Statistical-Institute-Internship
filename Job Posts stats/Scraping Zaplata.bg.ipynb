```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from dateutil import parser
pd.options.mode.chained_assignment = None  # default='warn'
```

### Paganation

```python
jobs_title_ = []
salaries_ = []
date_ = []
locations_ = []
companies_ = []

count = 1
while True:
    website_ = 'https://www.zaplata.bg/search/?q=&city=&city_distance=0&price=200%253B10000&cat%5B0%5D=1000&go=7&page=' + str(count)
    response_ = requests.get(website_)
    soup_ = BeautifulSoup(response_.content, 'html.parser')
    results_ = soup_.findAll('ul',{'class':"listItem"})
    
    if len(results_) == 0:
        break
    
    for result in results_:
        try:            
            date = result.find('span', {'class':'location'}).get_text().split()[0:2]
            date_.append(date)
        except:
            date_.append('n/a')           
            
        try:
            jobs_title_.append(result.find('li', {'class':'c2'}).find('a').get_text())
        except:
            jobs_title_.append('n/a')    
            
        try:
            salaries_.append(result.find('span', {'class':'is_visibility_salary'}).get_text())    
        except:
            salaries_.append('n/a')     
            
        try:
            companies_.append(result.find('li', {'class':'c4'}).get_text().strip()) 
        except:
            companies_.append('n/a') 
            
        try:
            locations_.append(result.find('span', {'class':'location'}).get_text().split(',')[1].strip())
        except:
            locations_.append('n/a')
                                                                                                                              
    count = count + 1                                                                                                                      
```

```python
df_jobs = pd.DataFrame({'Job_Title' : jobs_title_, 'Salary' : salaries_, 'Date' : date_,
                                     'Location' : locations_, 'Company' : companies_})
```

### Data Cleaning

```python
column_date = list(df_jobs['Date'])
column_date
```

```python
new_column_date = []
for item in column_date:
    day = str(item[0]).rstrip() 
    month = str(item[1]).replace(',','')
    if month == 'Януари':
        month = 'January'
    elif month == 'Февруари':
        month = 'February'
    elif month == 'Март':
        month = 'March'
    elif month == 'Април':
        month = 'April'
    elif month == 'Май':
        month = 'May'  
    elif month == 'Юни':
        month = 'June'  
    elif month == 'Юли':
        month = 'July' 
    elif month == 'Август':
        month = 'August' 
    elif month == 'Септември':
        month = 'September'   
    elif month == 'Октомври':
        month = 'October'   
    elif month == 'Ноември':
        month = 'November'    
    elif month == 'Декември':
        month = 'December'  
    new_column_date.append(f"{day}-{month}")
new_column_date
```

```python
df_jobs['Date'] = new_column_date
```

```python
df_jobs = df_jobs[~df_jobs.Date.str.contains("днес", na=False)]
df_jobs
```

```python
df_jobs['Date'] = [parser.parse(x) for x in df_jobs['Date']]
df_jobs
```

```python
df_jobs['Salary from'] = df_jobs['Salary'].str.extract('([0-9]+)')
df_jobs['Salary from'] = df_jobs['Salary from'].astype(float)
df_jobs.head()
```

```python
df_jobs['Salary to'] = df_jobs['Salary'].str.findall(r'([-+]?\d*\.\d+|\d+)').str[-1]
df_jobs['Salary to'] = df_jobs['Salary to'].astype(float)
df_jobs.head()
```

```python
df_jobs['Currency'] = df_jobs['Salary'].str.split().str[-1]
df_jobs.head()
```

```python
df_jobs.drop('Salary', axis=1, inplace=True)
df_jobs.head()
```

```python
# move column to specific place
column_to_move = df_jobs.pop("Date")
df_jobs.insert(0, "Date", column_to_move )
df_jobs.head()
```

```python
column_to_move = df_jobs.pop("Company")
df_jobs.insert(6, "Company", column_to_move )
df_jobs.head()
```

```python
df_jobs.Currency = df_jobs.Currency.replace('n/a', np.nan)
df_jobs.Currency = df_jobs.Currency.replace('лв.', 'BGN')
df_jobs.Currency = df_jobs.Currency.replace('евро', 'EUR')
df_jobs.head()
```

```python
df_jobs.drop_duplicates(keep='first', inplace=True, ignore_index=False)
df_jobs.info()
```

```python
df_jobs.to_excel('ads_Zaplata.bg.xlsx', index = False)
```
