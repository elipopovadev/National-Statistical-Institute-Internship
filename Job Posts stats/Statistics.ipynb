# What is the territorial distribution of job post?

```python
df_jobs.head()
```

```python
# get top 15 most frequent locations
n = 15
most_frequent_location = df_jobs['Location'].value_counts()[:n].index.tolist()
most_frequent_location
```

```python
temp_dict_location = {}
for town in most_frequent_location:
    count = df_jobs[df_jobs["Location"] == town]['Date'].count()
    temp_dict_location[town] = count
```

```python
# create DF from these most frequent locations
df_most_fr_location = pd.DataFrame(temp_dict_location,index=[0])
df_most_fr_location
```

```python
transform_df_location = df_most_fr_location.melt(var_name= "Територия", 
        value_name="Преброяване")
transform_df_location
```

```python
plt.figure(figsize=(15, 10), dpi=200)
x = transform_df_location['Територия']
y = transform_df_location['Преброяване']
plt.bar(x,y,align='center') # A bar chart
plt.xlabel('Територия')
plt.ylabel('Преброяване')
for i in range(len(y)):
    plt.hlines(y[i],0,x[i]) # Here you are drawing the horizontal lines
plt.title('Териториално разпределение на обявите за свободни работни места към 7 октомври 2022')
plt.xticks(rotation=45, ha='right')
plt.show()
```
